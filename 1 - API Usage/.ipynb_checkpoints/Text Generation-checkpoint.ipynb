```python
from transformers import pipeline, set_seed

import textwrap
import numpy as np
import matplotlib.pyplot as plt

from pprint import pprint
```

```python
with open("robert_frost.txt", "r") as f:
    for line in f.readlines():
        print(line)
```

```python
# Eliminiamo tutti gli spazi in eccesso alla destra di una stringa attraverso rstrip
# Dopo di che, siccome nel testo ci sono delle linee vuote, eliminiamo anche quelle
lines = [line.rstrip() for line in open("robert_frost.txt")]
lines = [line for line in lines if len(line) > 0]
```

```python
gen = pipeline("text-generation") # Default gpt2
```

```python
# Estendiamo la prima riga
gen(lines[0])
```
